```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.path import Path
import matplotlib.patches as patches

num_points = 15
x = np.random.uniform(1, 5, num_points)
y = np.random.uniform(1, 5, num_points)

fig, ax = plt.subplots()

ax.scatter(x, y, color='k')

fig_shape = Path([[1,2], [2,2], [2,1], [3,1], [3,2], [4,2],[4,3], [3,3], [3,4], [2,4], [2,3], [1,3], [1,2]])
patch = patches.PathPatch(fig_shape, facecolor= 'none', lw = 1)
ax.add_patch(patch)

for i in range(num_points):
  if fig_shape.contains_point((x[i], y[i])):
    ax.scatter(x[i], y[i], color='r')

ax.set_xlim(0, 5)
ax.set_ylim(0, 5)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.path import Path
import matplotlib.patches as patches

num_points = 15
x = np.random.uniform(1, 7, num_points)  # установка диапазона от 1 до 7
y = np.random.uniform(1, 7, num_points)  # установка диапазона от 1 до 7

fig, ax = plt.subplots()

ax.scatter(x, y, color='k')

# Создаем фигуру с использованием координат точек
fig_shape = Path([[1,2], [2,2], [2,1], [3,1], [3,2], [4,2], [3,3], [3,4], [2,4], [2,3], [1,3]])
patch = patches.PathPatch(fig_shape, facecolor='none', lw=1)
ax.add_patch(patch)

for i in range(num_points):
    if fig_shape.contains_point((x[i], y[i])):
        ax.scatter(x[i], y[i], color='r')  # перекрасить точку в красный, если она внутри фигуры

ax.set_xlim(0, 7)
ax.set_ylim(0, 7)

plt.show()
```
